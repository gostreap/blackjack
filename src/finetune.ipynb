```python
from q_learning import QLearning
from sarsa import Sarsa
from temporal_difference import TemporalDifference
```

```python
best_espilon, best_gamma, best_reward = None, None, -1000
for epsilon in [0.05, 0.1, 0.2, 0.3, 0.5]:
    for gamma in [1 / i for i in range(1, 8)]:
        print("#### {} {} ####".format(epsilon, gamma))
        model = QLearning("v1", epsilon=epsilon, gamma=gamma)
        model.train(40000)
        _, _, _, reward = model.test(10000)
        if reward > best_reward:
            best_reward = reward
            best_espilon = epsilon
            best_gamma = gamma
print(best_espilon, best_gamma, best_reward)
```
